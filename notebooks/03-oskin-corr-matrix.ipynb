Load data from excel file

```python
# Import modules
# import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from dython.nominal import associations

pd.set_option('display.max_columns', 100)
sns.set()
sns.set_style("white")
```

Read data from clean dataset

```python
df = pd.read_csv('../data/processed/wound_outcomes.csv')
```

```python
print(df.head())
```

#### Calculate the correlation/strength-of-association of features Cramer's V 

read more: https://towardsdatascience.com/the-search-for-categorical-correlation-a1cf7f1888c9

```python
corr=associations(df, nominal_columns='all', mark_columns=False, theil_u=False, plot=False, return_results=True)

plt.figure(figsize=(50,50))
plt.title = 'Correlation of features'
sns.heatmap(corr,annot=True,linewidth = 0.5, cmap='coolwarm', square=True)
```

Examine only features, having ``ИСХОД`` above zero in the correlation matrix

```python
corr_adj = corr.loc[corr['ИСХОД']>0, corr.loc['ИСХОД',:]>0]

plt.figure(figsize=(23,23))
plt.title = 'Correlation of features'
sns.heatmap(corr_adj,annot=True,linewidth = 0.5, cmap='coolwarm', square=True)
```
